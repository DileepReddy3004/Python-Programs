```python
def recursive_sum(*args):
    if len(args)==0:
        return 0
    if len(args)==1:
        return args[0]
    return args[0]+recursive_sum(*args[1:])
print(recursive_sum(10,20,30,40))
```

```python
students = [
    ("Ravi", 80),
    ("Amit", 70),
    ("Neha", 95),
    ("Sita", 60)
]

strategies = {
    "by_name": lambda x: x[0],
    "by_score": lambda x: x[1],
    "by_length": lambda x: len(x[0])
}

choice = input("Enter strategy (by_name/by_score/by_length): ")

sorted_students = sorted(students, key=strategies[choice])

print(sorted_students)
```

```python
students = [
    {"name": "Ravi", "score": 80},
    {"name": "Sita", "score": 55},
    {"name": "Amit", "score": 70},
    {"name": "Neha", "score": 40}
]

passed = filter(lambda x: x["score"] >= 60, students)

graded = map(lambda x: {**x, "grade": "Pass"}, passed)

result = sorted(graded, key=lambda x: x["score"], reverse=True)

print(result)
```
